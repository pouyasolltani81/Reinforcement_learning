```python
l = []
p_values = []
tensor_l = torch.Tensor([])
number_of_epochs = 20
number_of_batches = 0
sample_sizes = np.logspace(1, 7, num=50, dtype=int)

for i in tqdm.tqdm(sample_sizes):
    number_of_batches += 1
    xs = torch.rand(i, dtype=torch.double, device='cuda')
    ys = torch.rand(i, dtype=torch.double, device='cuda')
    # torch.cat([tensor_l, (xs.pow_(2) + ys.pow_(2) <= 1).sum()], dim=0)
    l.append(((xs.pow_(2) + ys.pow_(2) <= 1).sum() / (i) * 4).item())#torch.Tensor([i * 100_000_000]).to('cuda'))
    # tensor_l = torch.Tensor(l)
    # p_values.append((tensor_l.sum() * 4).item())
```

```python
fig, ax = plt.subplots()
ax.plot(sample_sizes, l, 'o-', color='xkcd:turquoise')
ax.axhline(np.pi, color='r', linestyle='--', label='True Value')
ax.set_xscale('log')
ax.set_xlabel('Number of Points')
ax.set_ylabel('Estimated Value of pi')
ax.set_title('Estimating pi using the Monte Carlo method')
ax.legend()
plt.show()
```

```python
pi_errors = np.abs(np.array(l) - np.pi)

# Plot the relationship between the error and the sample size
fig, ax = plt.subplots()
ax.plot(sample_sizes, pi_errors, 'o-', color='xkcd:salmon')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Sample Size')
ax.set_ylabel('Absolute Error in Estimate of pi')
ax.set_title('Error in Estimating pi using the Monte Carlo method')
plt.show()
```

```python
circle = plt.Circle((0, 0), 1, color='xkcd:coral', fill=False, linewidth=1, clip_on=False)

fig, ax = plt.subplots()
x_vals = []
y_vals = []
count = [0]

def animate(i):
    x_val = np.random.uniform(-1, 1)
    y_val = np.random.uniform(-1, 1)
    x_vals.append(x_val)
    y_vals.append(y_val)
    ax.scatter(0, 0, color='w', s=50, edgecolors='w')
    ax.cla()
    if i != 0:
        if i % 2 == 0:

            if (x_val**2 + y_val**2) <= 1:
                ax.scatter(x_val, y_val, color='xkcd:salmon', s=50, edgecolors='w')
            else:
                ax.scatter(x_val, y_val, color='xkcd:turquoise', s=50, edgecolors='w')
        else:
            ax.scatter(x_vals[-2], y_vals[-2], color='white', s=50, edgecolors='w')

    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    ax.add_patch(circle)
    
    # plt.xticks([], [])
    ax.set_xticks([], [])
    ax.set_xticks([])
    ax.set_xticks([], minor=True)

    # plt.yticks([], [])
    ax.set_yticks([], [])
    ax.set_yticks([])
    ax.set_yticks([], minor=True)

plt.close()
```

```python
ani = FuncAnimation(fig, animate, frames=1000, interval=200)
```

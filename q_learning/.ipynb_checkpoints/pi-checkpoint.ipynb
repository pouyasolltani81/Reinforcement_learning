# number of pi

To calculate the number of pi (π) and visualize it, we can use the Monte Carlo method, which is a statistical technique that relies on random sampling. Here’s how the method works:

Imagine a square with a side length of 2 units, inside which there is a circle with a radius of 1 unit.
Randomly generate points within the square.
Count how many points fall inside the circle (distance from the center is less than or equal to 1).
The ratio of the points inside the circle to the total number of points, multiplied by 4, gives an estimation of π.
This is because the area of the circle is πr² (π in this case, since r=1), and the area of the square is 4r² (4, since r=1). The ratio of the area of the circle to the square is π/4.

```python
import matplotlib.pyplot as plt
import numpy as np

# Set the number of random points
num_points = 20000

# Generate random points
x = np.random.rand(num_points)
y = np.random.rand(num_points)

# Calculate the distance from the origin (0,0)
distance_from_origin = np.sqrt(x**2 + y**2)

# Count the number of points within the unit circle
inside_circle = distance_from_origin <= 1

# Estimate Pi using the ratio of points inside the circle to the total number of points
estimated_pi = 4 * np.sum(inside_circle) / num_points

# Plot the points
plt.figure(figsize=(8, 8))
plt.scatter(x[inside_circle], y[inside_circle], color='blue', s=1, label='Inside Circle')
plt.scatter(x[~inside_circle], y[~inside_circle], color='red', s=1, label='Outside Circle')
plt.legend()

# Draw a unit circle for reference
circle = plt.Circle((0, 0), 1, color='green', fill=False)
plt.gca().add_artist(circle)

# Set the aspect of the plot to be equal
plt.axis('equal')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Estimation of Pi using Monte Carlo method: Pi ≈ {estimated_pi}')

# Show the plot

plt.show()
```

Another method to calculate the number of pi (π) is the Leibniz formula, which is a simple, yet slower method compared to others. The formula is given by: 
## π=4×k=0∑∞​2k+1(−1)k​=4×(1−31​+51​−71​+91​−...)
This series converges very slowly, but it’s a good example of a simple algorithm to calculate π. Here’s a Python code that uses the Leibniz formula to estimate π and visualizes the convergence:

```python

import numpy as np

# Leibniz formula for π
def calculate_pi_leibniz(num_terms):
    pi_estimate = 0
    for k in range(num_terms):
        pi_estimate += ((-1)**k) / (2*k + 1)
    pi_estimate *= 4
    return pi_estimate

# Set the number of terms for the approximation
num_terms = 1000

# Calculate π using the Leibniz formula
pi_values = [calculate_pi_leibniz(n) for n in range(1, num_terms + 1)]

# Visualize the convergence to π
plt.plot(pi_values, label='Estimate of π')
plt.axhline(np.pi, color='r', linestyle='--', label='Actual π')
plt.xlabel('Number of Terms')
plt.ylabel('Estimate of π')
plt.title(f'Convergence of Leibniz Formula for π. pi ≈ {pi_values[-1]}')
plt.legend()
plt.show()
```

Another method to calculate the number of pi (π) is the Nilakantha series, which is a more rapidly converging infinite series than the Leibniz formula. The Nilakantha series is given by:
## π=3+4×k=1∑∞​(2k)(2k+1)(2k+2)(−1)k+1​
This series starts with 3 and then continues with the sum of the reciprocals of the product of four consecutive numbers, with alternating signs.
Here’s a Python code snippet that uses the Nilakantha series to estimate π and visualizes the convergence:

```python

# Nilakantha series for calculating pi
def calculate_pi_nilakantha(terms):
    pi_estimate = 3
    sign = 1
    for i in range(2, 2 + terms * 2, 2):
        pi_estimate += sign * (4 / (i * (i + 1) * (i + 2)))
        sign *= -1
    return pi_estimate

# Set the number of terms for the approximation
num_terms = 1000

# Calculate pi using the Nilakantha series
pi_values = [calculate_pi_nilakantha(n) for n in range(1, num_terms + 1)]

# Visualize the convergence to pi
plt.plot(pi_values, label='Estimate of π')
plt.axhline(np.pi, color='r', linestyle='--', label='Actual π')
plt.xlabel('Number of Terms')
plt.ylabel('Estimate of π')
plt.title(f'Convergence of Nilakantha Series for π. pi ≈ {pi_values[-1]}')
plt.legend()
plt.show()
```
